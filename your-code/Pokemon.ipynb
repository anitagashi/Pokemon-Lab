```python
pokemon = pd.read_csv("Pokemon.csv")
```

# Print first 10 rows of pokemon

```python
pokemon.head(10)
```

# Extract values from Type 1 and Type 2

```python
type1 = pokemon['Type 1'].unique()
type2 = pokemon['Type 2'].unique()

distinct_values = set().union(type1, type2)

print(distinct_values)
```

# CleanUp names that contain Mega

```python
pokemon['Name'] = pokemon['Name'].str.replace('.*Mega', 'Mega')
pokemon.head(10)
```

# Calculate A/D Ratio

```python
pokemon['A/D Ratio'] = pokemon['Attack'] / pokemon['Defense']

pokemon.head()
```

# Find the pokemon with the highest A/D Ratio

```python
highest_ratio = pokemon['A/D Ratio'].idxmax()
pokemon_highest_ratio = pokemon.loc[highest_ratio, 'Name']
print("Pokémon with the highest A/D Ratio:", pokemon_highest_ratio)
```

# Find the pokemon with the lowest A/D Ratio

```python
lowest_ratio = pokemon['A/D Ratio'].idxmin()
pokemon_lowest_ratio = pokemon.loc[lowest_ratio, 'Name']
print("Pokémon with the lowest A/D Ratio:", pokemon_lowest_ratio)
```

# Create Combo Column

```python
pokemon['Combo Type'] = np.where(pokemon['Type 2'].isna(),pokemon['Type 1'], pokemon['Type 1']+'-'+pokemon['Type 2'])
```

```python
pokemon.head()
```

# Identify the pokemon whose A/D Ratio are among the top 5.

```python
top_5 = pokemon.nlargest(5,'A/D Ratio')
top_5
```

# For the 5 pokemon printed above, aggregate Combo Type and use a list to store the unique values.

```python
pokemon_sorted = pokemon.sort_values('A/D Ratio', ascending=False)

# Select the top 5 rows with the highest A/D Ratio
top_5_pokemon = pokemon_sorted.head(5)

# Aggregate Combo Type values and store unique values in a list
combo_types = list(set(top_5_pokemon['Combo Type']))

print(combo_types)
```

# For each of the Combo Type values obtained from the previous question, calculate the mean scores of all numeric fields across all pokemon.

```python
# your code here
mean_score = top_5.groupby('Combo Type').agg('mean')

# Print the mean scores
display(mean_score)
```

# Problem Solving Iteration 2

```python
import pandas as pd
pokemon2 = pd.read_csv('pokemon.csv')
```

```python
# Select the relevant columns for encoding
type_columns = ['Type 1', 'Type 2']
```

```python
# Apply One Hot Encoding
encoded_data = pd.get_dummies(pokemon2[type_columns], prefix=type_columns)
```

```python
# Concatenate the encoded data with the original dataset
encoded_pokemon2 = pd.concat([pokemon2, encoded_data], axis=1)
```

```python
display(encoded_pokemon2)
```

# Problem Solving Iteration 3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation of 'Total' with each encoded type field
correlations = encoded_pokemon2.corr()['Total'][2:]
```

```python
# Rank the correlations in descending order
ranked_correlations = correlations.sort_values(ascending=False)
```

```python
# Identify the Pokémon type with the highest correlation
top_type = ranked_correlations.index[0]
```

```python
# Print the result
print("The Pokémon type with the highest correlation to 'Total' is:", top_type)
```
